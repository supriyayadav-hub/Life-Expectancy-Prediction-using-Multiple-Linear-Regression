```python
#Since Hepatitis B, GDP, Population have more than 10% null values, delete the columns
lfc = lf.copy()
lfc.fillna(lfc.mean(), inplace = True)
```

```python
#Creatinng Dummy Variables
obj = lfc.dtypes==np.object
obj
```

```python
lfc.columns[obj]
```

```python
del lfc['Country']
```

```python
obj = lfc.dtypes == np.object
print(obj)
```

```python
lfc.columns[obj]
```

```python
dummydf = pd.DataFrame()
for i in lfc.columns[obj]:
    dummy = pd.get_dummies(lfc[i], drop_first = True)
    
    dummydf=pd.concat([dummydf, dummy], axis = 1)
print(dummydf)
```

```python
lfc1 = pd.concat([lfc, dummydf], axis = 1)
lfc1
```

```python
lfc
```

```python
del lfc1['Status']
lfc1
```

```python
#Applying Linear Regression Model
```

```python
x = lfc1.iloc[:,:-1].values
y = lfc1.iloc[:,-1].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split #importing split function
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 13)
```

```python
model = LinearRegression()
```

```python
model = model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
#Metrics
```

```python
print(model.intercept_)
```

```python
print(model.coef_)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
```

```python
from math import sqrt
rmse = sqrt(mse)
```

```python
print("Mean Squared Error : ", mse)
print("Root Mean Squared Error : ", rmse)
```
